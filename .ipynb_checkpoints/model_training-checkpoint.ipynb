```python
class BertForSequenceClassification(nn.Module):
    def __init__(self, model_name, num_labels):
        super(BertForSequenceClassification, self).__init__()
        self.bert = AutoModel.from_pretrained(model_name)
        self.classifier = nn.Linear(self.bert.config.hidden_size, num_labels)
        self.class_weights = class_weights
        self.f1 = F1Score(num_classes=num_labels, task='multiclass')
        
    def forward(self, input_ids, attention_mask=None, labels=None):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        logits = self.classifier(outputs.last_hidden_state[:, 0, :])  
        
        loss = None
        f1_score = None
        if labels is not None:
            #loss_fn = nn.CrossEntropyLoss(weight=self.class_weights)
            #loss = loss_fn(logits, labels)
            f1_score = self.f1(logits.argmax(dim=1), labels)
        wandb.log({"f1_score": f1_score, 
                   #'CrossEntropyLoss':loss
                  }
                 )
        return (loss, logits, f1_score) if loss is not None else (logits, f1_score)
```

# Modello addestrato sui codici delle azioni

```python
df = pd.read_parquet('dataset_BERT/dataset_atemporale.gzip').rename(columns={'ID_tassonomia':'label'})
df = df[['text', 'label']]
df.label = df.label - 1
#df = df.sample(100, random_state=25)
num_labels = df.label.unique().shape[0]

t1, test_set = train_test_split(df[['text', 'label']], test_size=0.2)
train_set, val_set =  train_test_split(t1, test_size=0.15)
unique_labels, conteggi_classi = np.unique(df.label, return_counts=True)
unique_labels, conteggi_classi 
inverse = 1/conteggi_classi 
normalizzati = inverse / inverse.sum()  

class_weights = torch.FloatTensor(normalizzati).to('cuda')
dataset = DatasetDict({
    'train': Dataset.from_pandas(train_set),
    'validation': Dataset.from_pandas(val_set),
    'test': Dataset.from_pandas(test_set)
})

model_name = "dbmdz/bert-base-italian-xxl-cased"

#model_name = "dbmdz/bert-base-italian-cased"

tokenizer = AutoTokenizer.from_pretrained(model_name, 
                                          max_position_embeddings=512, 
                                          use_cache=False)
def tokenize_function(examples):
    return tokenizer(examples['text'], padding=True, truncation=True, return_tensors="pt")

tokenized_datasets = dataset.map(tokenize_function, batched=True).shuffle(seed=25)
tokenized_datasets = tokenized_datasets.remove_columns(['text', '__index_level_0__', 'token_type_ids'])
```

```python
model = BertForSequenceClassification(model_name, num_labels)
transformers.logging.set_verbosity_info()

training_args = TrainingArguments(
    output_dir='./results',
    eval_strategy='epoch',
    save_total_limit=5,
    # parametri per la lunghezza del training
    num_train_epochs=1,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    do_train=True,
    gradient_accumulation_steps=2
    #max_steps=50,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['validation'],
    
)
torch.cuda.empty_cache()
trainer.train()
trainer.evaluate()
```

# Modello addestrato sui macroambiti

## Dati

```python
df = pd.read_parquet('dataset_BERT/dataset_atemporale.gzip').rename(columns={'macroambito':'label'})
df = df[['text', 'label']]
mappatura = {'A':0,'B':1,'C':2,'D':3,'E':4}
df.label = df.label.apply(lambda x: mappatura[x])
#df = df.sample(100, random_state=25)
```

```python

num_labels = df.label.unique().shape[0]
t1, test_set = train_test_split(df[['text', 'label']], test_size=0.2)
train_set, val_set =  train_test_split(t1, test_size=0.15)
```

```python

unique_labels, conteggi_classi = np.unique(train_set.label, return_counts=True)
unique_labels, conteggi_classi 
inverse = 1/conteggi_classi 
normalizzati = inverse / inverse.sum()  

class_weights = torch.FloatTensor(normalizzati).to('cuda')
```

```python

dataset = DatasetDict({
    'train': Dataset.from_pandas(train_set),
    'validation': Dataset.from_pandas(val_set),
    'test': Dataset.from_pandas(test_set)
})

model_name = "dbmdz/bert-base-italian-xxl-cased"

tokenizer = AutoTokenizer.from_pretrained(model_name, 
                                          max_position_embeddings=1024, 
                                          use_cache=False)
def tokenize_function(examples):
    return tokenizer(examples['text'], padding=True, truncation=True, return_tensors="pt")

tokenized_datasets = dataset.map(tokenize_function, batched=True).shuffle(seed=25)
tokenized_datasets = tokenized_datasets.remove_columns(['text', '__index_level_0__', 'token_type_ids'])
```

```python
tokenized_datasets
```

## Training

```python
torch.cuda.empty_cache()
model = BertForSequenceClassification(model_name, num_labels)
transformers.logging.set_verbosity_info()

training_args = TrainingArguments(
    output_dir='./results',
    eval_strategy='epoch',
    remove_unused_columns=True,
    dataloader_num_workers=2,
    save_total_limit=5,
    
    # parametri per la lunghezza del training
    do_train=True,
    num_train_epochs=6,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    gradient_accumulation_steps=2
    #max_steps=50,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['validation'],
    
)

trainer.train()
```

## Valutazione

```python
predictions = trainer.predict(tokenized_datasets['test'])
```

```python
predictions.predictions[0]
```

```python
logits = torch.tensor(predictions.predictions[0])
labels = torch.tensor(predictions.label_ids)

# Compute the F1 score
f1 = F1Score(num_classes=num_labels, task='multiclass')
f1_score = f1(logits.argmax(dim=1), labels)

print(f"F1 score: {f1_score.item()}")
```
